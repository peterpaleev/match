```python
import math
import matplotlib.pyplot as plt

# True value of sin(1)
true_sin_1 = math.sin(1)

def calculate_accurate_digits(approx_value, true_value):
    error = abs(approx_value - true_value)
    error_str = f"{error:.50f}"
    accurate_digits = 0
    for i in error_str:
        if i == '0' or i == '.':
            accurate_digits += 1
        else:
            break
    return accurate_digits

# List to store the number of accurate digits per iteration
accurate_digits_list = []

# Number of iterations
n_iterations = 100

for num_terms in range(1, n_iterations + 1):
    sin_1_approx = sum((-1)**n * (1**(2*n + 1)) / math.factorial(2*n + 1) for n in range(num_terms))
    digits = calculate_accurate_digits(sin_1_approx, true_sin_1)
    accurate_digits_list.append(digits)
    print(f"Terms: {num_terms}, Approximation: {sin_1_approx}, Digits: {digits}")

# Print the list to verify the content
print(accurate_digits_list)

# Plotting the results
plt.figure(figsize=(10, 5))  # Adjust the size to ensure all data is visible
plt.plot(range(1, n_iterations + 1), accurate_digits_list)
plt.xlabel('Number of Terms')
plt.ylabel('Number of Accurate Digits')
plt.title('Accuracy of Taylor Series Approximation for sin(1)')
plt.grid(True)
plt.show()
```
